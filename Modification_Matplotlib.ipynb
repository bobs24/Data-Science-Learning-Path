Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
#code here
#plt.figure(figsize=(20,20))
fig,new=plt.subplots()
new.plot(x,cos,color='Red',label='Cosx')
new.plot(x,cos3,linestyle='--',color='Green',label='3 Cosx')
new.plot(x,sin,linestyle='dashdot',color='Blue',label='Sin(x-90)')
new.plot(x,sin2,marker='x',linestyle=' ',color='Yellow',label='2 Sinx')
new.set_xlabel('X-Axis Label')
new.set_ylabel('Y-Axis Label')
new.set_title('Grafik Cos Sin')
plt.legend(loc='center')
#plt.axis([-10,10,-3,3])
plt.show()
```
