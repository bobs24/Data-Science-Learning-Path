<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,squeeze=False,sharey=True,figsize=(14,7))
ax[0,0].plot(x,tan,color='Blue',label='tanX')
ax[0,0].set_title('Grafik tanx')
ax[0,1].plot(x,sin,color='Red',label='sinX')
ax[0,1].set_title('Grafik sinx')
ax[0,2].plot(x,cos,color='Green',label='cosX')
ax[0,2].set_title('Grafik cosx')
fig.legend(loc='upper center')
plt.tight_layout(6)
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(12,10))
ax.plot(x,y)

ins1 = ax.inset_axes([0.867,0.9,0.4,0.3])
ins1.plot(x, y4,color='Black')
ins1.plot(x, y,color='Blue')
ins1.set_title('Grafik CosX^4')

ins2 = ax.inset_axes([-0.133, 0.9, 0.4, 0.3])
ins2.plot(x, y5,color='Orange')
ins2.plot(x, y,color='Blue')
ins2.set_title('Grafik CosX^5')

ins3 = ax.inset_axes([0.867, -0.1, 0.4, 0.3])
ins3.plot(x, y3,color='Green')
ins3.plot(x, y,color='Blue')
ins3.set_title('Grafik CosX^3')

ins4 = ax.inset_axes([-0.133,-0.1 , 0.4, 0.3])
ins4.plot(x, y2,color='Red')
ins4.plot(x, y,color='Blue')
ins4.set_title('Grafik CosX^2')

fig.suptitle('Grafik CosX')
plt.show()
```
