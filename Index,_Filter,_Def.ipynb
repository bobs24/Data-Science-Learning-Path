<h2>Soal 2: Filter DataFrame Dengan Boolean</h2>

1. Tampilkan semua data siswa yang level pendidikan orang tuanya 'high school' dan 'bachelor degree'
2. Tampilkan semua data siswa yang mempunyai nilai lebih dari 50 dengan test preparation course 'completed'
3. Tampilkan semua data siswa yang mempunyai nilai lebih dari 70, berjenis kelamin perempuan dengan test preparation course 'completed'

```python
#code here
data_siswa = df["parental level of education"] =='high school'
data_siswa1 = df["parental level of education"] =='bachelor degree'
df[(data_siswa)|(data_siswa1)]
```

```python
#Tampilkan semua data siswa yang mempunyai nilai lebih dari 50 dengan test preparation course 'completed'
data_siswa = df["nilai"]>50
data_siswa1 = df["test preparation course"] =='completed'
df[(data_siswa)&(data_siswa1)]
```

```python
#Tampilkan semua data siswa yang mempunyai nilai lebih dari 70, berjenis kelamin perempuan dengan test preparation course 'completed'
data_siswa = df["nilai"]>70
data_siswa1 = df["gender"] =='female'
data_siswa2 = df["test preparation course"] =='completed'
df[(data_siswa)&(data_siswa1)&(data_siswa2)]
```

<h2>Soal 3: Data Transformation</h2>

lakukan data transformasi terhadap column 'nilai' dengan fungsi transformasi, berikut rumus dari normalisasi

![alt text](https://drive.google.com/uc?id=1wed-EbRSTA0JP1ceDp6lcKBqH2yJ4aa3)

```python
#code here
def minimum_minus(nilai):
  return nilai-nilai.min()
def max(nilai):
  return nilai.max()
def min(nilai):
  return nilai.min()
print((df[['nilai']].apply(minimum_minus))/((df[['nilai']].apply(max))-(df[['nilai']].apply(min))))
```
